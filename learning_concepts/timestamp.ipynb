```python
from pprint import pprint
from typing import List, Optional

tts_word_timings = [ 
    {"index": 0, "start": 3.42, "end": 3.51, "word": "a"},
    {"index": 1, "start": 3.51, "end": 3.87, "word": "pool"}, 
    {"index": 2, "start": 3.87, "end": 4.11, "word": "buts"},
    {"index": 3, "start": 4.11, "end": 4.47, "word": "clever"},
    {"index": 4, "start": 4.47, "end": 4.77, "word": "man"},
    {"index": 5, "start": 4.77, "end": 5.13, "word": "named"},
    {"index": 6, "start": 5.13, "end": 5.30, "word": "milk"},
    {"index": 7, "start": 5.37, "end": 5.67, "word": "to"}, 
]

correct_word_timings = [
    {"word": "a"},
    {"word": "poor"},
    {"word": "but"},
    {"word": "clever"},
    {"word": "man"},
    {"word": "named"},
    {"word": "mircho"},
    {"word": "he"},
]

def getMatchedPos(words: List[str], startingFrom: int, word: str) -> int:
    try:
        return words.index(word, startingFrom)
    except ValueError:
        return -1

def align_timings(tts_word_timings: List[dict], correct_word_timings: List[dict]) -> List[dict]:
    tts_words = [w["word"] for w in tts_word_timings]
    final = []
    j = 0  # pointer for tts_word_timings
    last_matched_end = 0.0

    for i, correct in enumerate(correct_word_timings):
        correct_word = correct["word"]
        match_index = getMatchedPos(tts_words, j, correct_word)

        if match_index != -1:
            # Handle any missed words before this match
            missing = []
            for k in range(i - 1, -1, -1):
                if correct_word_timings[k]["word"] not in [f["word"] for f in final]:
                    missing.insert(0, correct_word_timings[k]["word"])
                else:
                    break

            # Determine time gap for missing words
            current_start = tts_word_timings[match_index]["start"]
            gap = current_start - last_matched_end

            if gap > 0 and missing:
                per = gap / len(missing)
                for idx_m, mword in enumerate(missing):
                    start = round(last_matched_end + idx_m * per, 3)
                    end = round(start + per, 3)
                    final.append({
                        "index": None,
                        "start": start,
                        "end": end,
                        "word": mword
                    })
                last_matched_end = final[-1]["end"]

            elif gap <= 0 and missing and final:
                prev = final[-1]
                total = prev["end"] - prev["start"]
                split = total / (len(missing) + 1)
                prev["end"] = round(prev["start"] + split, 3)
                for idx_m, mword in enumerate(missing):
                    start = round(prev["start"] + split * (idx_m + 1), 3)
                    end = round(start + split, 3)
                    final.append({
                        "index": None,
                        "start": start,
                        "end": end,
                        "word": mword
                    })
                last_matched_end = final[-1]["end"]

            # Append the matched word
            matched = tts_word_timings[match_index]
            final.append({
                "index": i,
                "start": matched["start"],
                "end": matched["end"],
                "word": correct_word
            })
            last_matched_end = matched["end"]
            j = match_index + 1

        else:
            # No match found at all
            duration = 0.5
            start = round(last_matched_end, 3)
            end = round(start + duration, 3)
            final.append({
                "index": i,
                "start": start,
                "end": end,
                "word": correct_word
            })
            last_matched_end = end
            j += 1

    return final

# Usage
result = align_timings(tts_word_timings, correct_word_timings)
pprint(result)
```

```python
tts_word_timings = [
    {"index": 1, "start": 0.0, "end": 0.3, "word": "the"},
    {"index": 2, "start": 0.3, "end": 0.6, "word": "last"},
    {"index": 3, "start": 0.6, "end": 0.9, "word": "word"}
]

correct_word_timings = [
    {"word": "the"},
    {"word": "very"},
    {"word": "very"},
    {"word": "last"},
    {"word": "word"}
]

# Build a quick lookup: correct_idx -> TTS timing or None
tts_map = {t["word"]: t for t in tts_word_timings}
final = []
last_matched_end = 0.0

for idx, cw in enumerate(correct_word_timings):
    w = cw["word"]
    if w in tts_map:
        # Before appending this matched word, insert any previous missing words
        # Find how many missing were immediately before this one
        prev_idx = len(final)  # number of already placed
        # Extract those missing
        missing = []
        # Look back in correct_word_timings between last placed index and this idx
        placed = {item["word"] for item in final}
        for back in range(idx - 1, -1, -1):
            if correct_word_timings[back]["word"] not in placed:
                missing.append(correct_word_timings[back]["word"])
            else:
                break
        missing.reverse()

        # Calculate the gap BEFORE this matched word
        next_start = tts_map[w]["start"]
        gap = next_start - last_matched_end

        if gap <= 0 and missing:
            # Zeroâ€‘width gap: re-split the previous segment instead
            prev = final[-1]
            seg_start, seg_end = prev["start"], prev["end"]
            total = seg_end - seg_start
            # first half stays, second half for first missing
            half = total / (len(missing) + 1)
            # adjust the previous word
            prev["end"] = round(seg_start + half, 3)
            last_matched_end = prev["end"]
            # assign missing words on subsequent halves
            for i_m, mw in enumerate(missing):
                start = round(seg_start + half * (i_m + 1), 3)
                end   = round(start + half, 3)
                final.append({"index": None, "start": start, "end": end, "word": mw})
            # now last_matched_end moves to end of last missing
            last_matched_end = final[-1]["end"]
        else:
            # Normal positive gap: split evenly among missing
            if missing:
                per = gap / len(missing)
                for i_m, mw in enumerate(missing):
                    start = round(last_matched_end + per * i_m, 3)
                    end   = round(start + per, 3)
                    final.append({"index": None, "start": start, "end": end, "word": mw})
            last_matched_end = next_start

        # Finally, add this matched word
        mat = tts_map[w]
        final.append({"index": idx, "start": mat["start"], "end": mat["end"], "word": w})
        last_matched_end = mat["end"]

# Print the result
from pprint import pprint
pprint(final)
```

```python
import requests
import logging
import os

def images_generates(prompt: str, save_path: str = "generated_image.png"):
    try:
        response = requests.post(
            "https://47ba-34-82-239-141.ngrok-free.app/generate",
            json={"prompt": prompt},
            timeout=60
        )
        result = response.json()
        if "image_url" not in result:
            raise ValueError("Image URL not returned from API")

        image_url = result["image_url"]
        image_data = requests.get(image_url).content

        # Save to local PC
        output_dir = os.path.dirname(save_path)
        if output_dir and not os.path.exists(output_dir):
            os.makedirs(output_dir)
        with open(save_path, "wb") as f:
            f.write(image_data)

        print(f"Image saved to {save_path}")
        return save_path

    except Exception as e:
        logging.error(f"Error calling Colab image generation: {e}")
        raise
images_generates("Generate an image of a cat sitting on Mars", "local_output/cat_on_mars.png")
```

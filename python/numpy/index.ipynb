```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

# Marks data
marks = [90, 85, 80, 85, 75, 80, 70, 60, 55, 20]

# Calculate statistics
mean_val = np.mean(marks)
median_val = np.median(marks)

# Create intervals (bins) from 0 to 100 in steps of 10
bins = np.arange(0, 110, 10)

# Plot histogram with density=True to overlay the curve
counts, _, _ = plt.hist(marks, bins=bins, edgecolor='black', alpha=0.6, density=True, label='Histogram')

# Create a smooth curve using KDE
kde = gaussian_kde(marks)
x_vals = np.linspace(0, 100, 500)
y_vals = kde(x_vals)

# Plot the smooth curve
plt.plot(x_vals, y_vals, color='red', linewidth=2, label='Distribution Curve')

# Highlight mode (most frequent bin manually identified as ~85)
plt.axvline(85, color='green', linestyle='--', label='Mode (~85)')

# Highlight mean and median
plt.axvline(mean_val, color='blue', linestyle='--', label=f'Mean ({mean_val:.1f})')
plt.axvline(median_val, color='purple', linestyle='--', label=f'Median ({median_val:.1f})')

# Labels and grid
plt.title("Negatively Skewed Distribution of Student Marks")
plt.xlabel("Marks")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde

# Positively skewed marks data (low scores are common, few high scores)
marks = [10, 15, 20, 20, 25, 30, 35, 40, 90, 95]

# Calculate statistics
mean_val = np.mean(marks)
median_val = np.median(marks)

# Create intervals (bins) from 0 to 100 in steps of 10
bins = np.arange(0, 110, 10)

# Plot histogram with density=True to overlay the curve
counts, _, _ = plt.hist(marks, bins=bins, edgecolor='black', alpha=0.6, density=True, label='Histogram')

# Create a smooth curve using KDE
kde = gaussian_kde(marks)
x_vals = np.linspace(0, 100, 500)
y_vals = kde(x_vals)

# Plot the smooth curve
plt.plot(x_vals, y_vals, color='red', linewidth=2, label='Distribution Curve')

# Highlight mode (most frequent bin manually identified as ~20)
plt.axvline(20, color='green', linestyle='--', label='Mode (~20)')

# Highlight mean and median
plt.axvline(mean_val, color='blue', linestyle='--', label=f'Mean ({mean_val:.1f})')
plt.axvline(median_val, color='purple', linestyle='--', label=f'Median ({median_val:.1f})')

# Labels and grid
plt.title("Positively Skewed Distribution of Student Marks")
plt.xlabel("Marks")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data: Marks of 10 students
marks = [20, 60, 65, 70, 72, 75, 78, 80, 82, 85]

# Create bins (intervals) of 10 from 0 to 90
bins = np.arange(0, 100, 10)  # 0-10, 10-20, ..., 90-100

# Calculate statistics
mean = np.mean(marks)
median = np.median(marks)
mode = max(set(marks), key=marks.count)

# Plot histogram
plt.figure(figsize=(10, 6))
counts, bins, patches = plt.hist(marks, bins=bins, edgecolor='black', alpha=0.7, color='skyblue')

# Plot vertical lines for mean, median, and mode
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.1f}')
plt.axvline(median, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median}')
plt.axvline(mode, color='blue', linestyle='dashed', linewidth=2, label=f'Mode: {mode}')

# Add labels and title
plt.title('Negatively Skewed Distribution of Student Marks (Binned by 10)')
plt.xlabel('Marks')
plt.ylabel('Number of Students')
plt.xticks(bins)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import uuid
from typing import Literal
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_google_genai import ChatGoogleGenerativeAI 
 
model = ChatGoogleGenerativeAI(model='gemini-1.5-flash', api_key=GEMINI_API_KEY)
 
def update_memory(user_info: str, todo_info: str, instruction_info: str) -> str:
    """
    Updates memory for user profile, todo list, and instructions.
    
    Args:
        user_info: The user's personal introduction.
        todo_info: Details about a to-do item.
        instruction_info: Guidelines or preferences.
    
    Returns:
        A confirmation message.
    """
    response_parts = []
    if user_info:
        response_parts.append(f"User info updated: {user_info}")
    if todo_info:
        response_parts.append(f"Todo updated: {todo_info}")
    if instruction_info:
        response_parts.append(f"Instructions updated: {instruction_info}")
    return " | ".join(response_parts)
 
llm_with_tools = model.bind_tools([update_memory])
 
MODEL_SYSTEM_MESSAGE = """You are a helpful assistant that updates user memory.
When you receive a user's message, analyze it and extract:
- The user's personal introduction (if present)
- Details about a to-do item (if present)
- Any instructions or guidelines (if present)

Then call the update_memory function with three arguments:
- user_info: the user's personal introduction (or an empty string if not present)
- todo_info: the to-do details (or an empty string if not present)
- instruction_info: any instructions or guidelines (or an empty string if not present)

For example, if the message is:
  "hi i am arman tomarrow i will go to lahore please use local vendors for swim lessons"
then:
  user_info should be "hi i am arman"
  todo_info should be "tomarrow i will go to lahore"
  instruction_info should be "please use local vendors for swim lessons".
"""

def task_mAIstro():
    system_msg = SystemMessage(content=MODEL_SYSTEM_MESSAGE) 
    profile = HumanMessage(content="hi i am arman tomarrow i will go to lahore please use local vendors for swim lessons, who is allama iqbal")
    final_resp = llm_with_tools.invoke([system_msg, profile])
    print(f'final_resp: {final_resp}')

task_mAIstro()
```

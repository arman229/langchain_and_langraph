```python
import os 
from dotenv import load_dotenv
load_dotenv()

GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")

if not GEMINI_API_KEY:
    raise ValueError("GEMINI_API_KEY is not set in the.env file.")

print("API key loaded from.env file.")
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI 
llm:ChatGoogleGenerativeAI = ChatGoogleGenerativeAI(api_key= GEMINI_API_KEY,model='gemini-1.5-flash',temperature=0.3)
```

```python
from langchain.prompts import PromptTemplate
def generate_quiz_questions(content,num_mcqs,num_short_questions):

    MIN_CONTENT_LENGTH = 500
    if len(content.strip()) < MIN_CONTENT_LENGTH:
        return f"Content provided is too short. Please provide more content with at least {MIN_CONTENT_LENGTH} characters."
 
    MAX_CONTENT_LENGTH = 2000   
    content = content.strip()[:MAX_CONTENT_LENGTH]   
      
    prompttemplate = PromptTemplate(
        input_variables=['context', 'num_mcqs', 'num_short_questions','formate_temp'],
        template="""
    You are a professional quiz creator. Your task is to generate two types of questions from the following content:

    1. **Multiple-Choice Questions (MCQs)**:
      - Generate exactly {num_mcqs} MCQs.
      - Each question should have 4 choices (a, b, c, d), with one correct answer.
      - Ensure the questions cover the most important topics and ideas in the content.
      - Format each MCQ as follows:
          {formate_temp}

    2. **Short Questions**:
      - Generate exactly {num_short_questions} short, clear, and concise questions.
      - Focus only on direct, factual, or conceptual questions that require short answers.
      - Ensure the questions cover the most important topics and ideas in the content.
      - Format each short question as follows:
          Q1. [Short question here]
          Q2. [Short question here]
          ...
          Q{num_short_questions}. [Short question here]

    Here is the content:

    {context}

    Generate the questions in the specified formats and ensure the output is structured as follows:
    - MCQs: A JSON array of MCQ objects.
    - Short Questions: A numbered list of short questions.
    """
    )
    formate = """
      [
        {
            "question": "The question text",
            "choices": [
                "Choice text for a",
                "Choice text for b",
                "Choice text for c",
                "Choice text for d"
            ],
            "correct_answer": "The actual text of the correct choice"
        }
      ]
    """
    
    formatted_prompt = prompttemplate.format(context=content,
    num_mcqs=num_mcqs,
    num_short_questions=num_short_questions,formate_temp=formate)
    try:
      final_result = llm.invoke(formatted_prompt)
      return final_result
    except Exception as e:
      return f"Error: {e}"
     
```

```python
 
import json
mcqs = []
short_questions = [] 
try:
    mcqs_start = final_resp.find("[")
    mcqs_end = final_resp.rfind("]") + 1   
    mcqs_json = final_resp[mcqs_start:mcqs_end]
    mcqs = json.loads(mcqs_json)   
except json.JSONDecodeError as e:
    print(f"Error parsing MCQs JSON: {e}")
 
short_questions_start = final_resp.find("Q1.")
if short_questions_start != -1:
    short_questions_text = final_resp[short_questions_start:].strip()
    short_questions = short_questions_text.split("\n")  
  
 
quiz_data = {
    "mcqs": mcqs,
    "short_questions": short_questions
}
 
print("quiz_data:", quiz_data)
 
```
